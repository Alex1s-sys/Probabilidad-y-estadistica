9. Diagrama de Cajas y bigotes
1. Las la variabilidad fuera de los cuartiles fore inferior
2. Los valores atipicos se representan bigotes como puntos individuales que están en linea con los
3. Si los datos simétricos, la mediana
4. Si los datos están sesgados, la mediana estará más cerca de la parte superior o inferior de la caja
5. Los entremos de arriba y abajo de la caja indican los cuantiles, o percentiles, 25 y 75. Estos dos cuantiles también se conocen como cuartiles, porque separan cuartos (25%) de los datos. La longitud de la caja es la diferencia entre estos dos percentiles y se conoce como rango intercuartilico (IQR)
6. Los diagramas caja-bigotes son muy útiles para comparar dos o más conjuntos de datos.



```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(40, 93), 2)
    fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
plt.boxplot(fabrica1)
plt.ylabel('Produccion en mts.')
plt.title('Produccion de la fabrica 1')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
vert=True, # Alineacion Vertical
patch_artist=True) # Rellenar con color
ax1.set_title('Producción fabrica 1')
bplot2 = ax2.boxplot(fabrica2,
vert=True,
# Alineacion Vertical
patch_artist=True) # Rellenar con color
ax2.set_title('Produccion fabrica 2')
plt.show()
```

```python
#Insertamos un valor de 150 a la fabrica 1,
# Seira un valor atipico
fabrica1.append(150)
print(fabrica1)
```

```python
fig, (ax1, ax2) = plt.subplots (nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1,
vert=True, # vertical box alignment
patch_artist=True) # fill with color
ax1.set_title('Producción fabrica 1')
bplot2 = ax2.boxplot(fabrica2,
vert=True, # vertical box alignment
patch_artist=True) # fill with color
ax2.set_title('Producción fabrica 2')
plt.show()
```

```python
#Agregamos ax1.set_ylim(8, 168) para garantizar que ambos gráficos usen La
#misma escala vertical, facilitando la comparación visual.
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))
bplot1 = ax1.boxplot(fabrica1,
vert=True, #Alineacion Vertical t
patch_artist=True) # Rellenar con color
ax1.set_title('Producción fabrica 1')
ax1.set_ylim(30, 160) # Escala Uniforme de 8 a 160
bplot2 = ax2.boxplot(fabrica2,
vert=True,#Alineacion Vertical t
patch_artist=True) # Rellenar con color
ax2.set_title('Producción fabrica 2')
ax2.set_ylim(30, 160) # Escala Uniforme de 8 a 160
plt.show()
```
