```python
#Basado en el libro estadisitica para la administracion pagina 12
#creamos la lista vacia que representa la edad de 100 personas
import random
edades = []
#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  #Generamos un numero aleatorio con la funcion randint y lo guardamos en una variable #numero aleatorio
  numero_aleatorio = random.randint(18, 70)
  #Agregamos el numero aleatorio a la lista edades
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#Crear un Dataframe de padas con los numero aleatorios
df=pd.DataFrame(edades,columns=['Edad'])
df.head()
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias=df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns=['Edad','Frecuencia']
tabla_frecuencias=tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla de distribucion de frecuencias
print("Tabla de distribucion de Frecuencias: \n ")
tabla_frecuencias
```

```python
tabla_frecuencias['Frecu relativa']= tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']= tabla_frecuencias['Frecuencia']/ 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecuencia acomulada']= tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

```python
#Importamos la librerias de funciones matematicas
import math
```

```python
#Volvemos a usar la lista de edades
print(edades)
```

```python
#Creamos un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
#Calcuar el numero de intervalos usando la regla de Struges
n= len(edades)
k=int(1+ math.log2(n))
print("Valor de n=", n ,"\nNumero de intervalos=", k)
```

```python
#Crear la tabla de frecuencia
tabla_frecuencia= pd.cut(df['Edad'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns=['Clase', 'Frec']
tabla_frecuencia
```

```python
tabla_frecuencia['Frec relativa']= tabla_frecuencia['Frec']/n
tabla_frecuencia['Frexx Porc']= tabla_frecuencia['Frec']/ n * 100
tabla_frecuencia
#
```

```python
#Ordena la tabla de frecuencia por el intervalo
tabla_frecuencia= tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
#Generar columnas siguientes
tabla_frecuencia['Frec Rel']=tabla_frecuencia['Frec']/n
tabla_frecuencia['Frec Porc']= tabla_frecuencia['Frec']/ n * 100
tabla_frecuencia['Frec Az']=tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac']=tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac']=tabla_frecuencia['Frec Porc'].cumsum()

#Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```
